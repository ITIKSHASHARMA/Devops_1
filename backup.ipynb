```python
# backup with compress
import shutil
import datetime
import os
```

```python
today= datetime.date.today()
```

```python
def backup_file(source,destination):
  backup_file_name= os.path.join(destination,f"backup_{today}")
  shutil.make_archive(backup_file_name,'gztar',source)
```

```python
source = ''
destination =''
```

```python
backup_file(source,destination)
```

```python
#script to backup from local to aws s3 use boto3 library
import boto3
```

```python
s3=boto3.resource("s3")
```

```python
bucket_name= "python-for-devops"
region= "us-east-2"
```

```python
def show_buckets(s3): #show all bucket list
  for bucket in s3.buckets.all():
    print(bucket.name)
show_buckets(s3)
```

```python
def create_bucket(s3,bucket_name,region):
  s3.create_bucket(bucket=bucket_name,CreateBucketConfiguration={
        'LocationConstraint': region,
    },)
  print("bucket create successfully")
create_bucket(s3)
```

```python
def upload_backup(s3,file_name,bucket_name,key_name):
  data=open(file_name,'rb') #read data in binary
  s3.bucket(bucket_name).put_object(key=key_name,Body=data) #insert data in body
  print("backup uploaded successfully")
```
